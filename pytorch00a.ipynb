## Reshaping , Stacking , Squizing , Unsquizing

```python
z = torch.arange(1. , 10.)
z , z.shape , z.ndim
```

```python
# Add an extra dimension
z_reshaped = z.reshape(1 , 9)
z_reshaped , z_reshaped.shape , z.ndim , z_reshaped.ndim
```

```python
# Change in view
a = z.view(1 , 9)
a , a.shape

# a and z share the same memory , means changes made in a will make changes in z
```

## Reproducability (create random out of random)

```python
torch.rand(3,3)
```

```python
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)
print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# Lets make some ram=ndom but reproducable tensors
import torch

# Set the random tensor
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3,4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3,4)
print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
